Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

The original frequencies in the example were [0, 1, 2, 3, 4, -3, -2, -1]. I will mltiply all the frequencies by 2 to make the array [0, 2, 4, 6, 8, -6, -4, -2].

```python
N = 8
ts = numpy.arange(N) / N
freqs = numpy.arange(N)
freqs = [0, 2, 4, 6, 8, -6, -4, -2]
args = numpy.outer(ts, freqs)
M2 = numpy.exp(i * PI2 * args)
M2
```

## Question 2

 In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

We will start with a high pass filter. I begin by creating a square wave and applying a high pass filter to its spectrum.

```python
sig = thinkdsp.SquareSignal()
sig.plot()
```

```python
wave = sig.make_wave(duration=1, framerate=44100)
spec = wave.make_spectrum()
```

```python
spec2 = wave.make_spectrum()
spec2.high_pass(5000)
spec2.plot(0.5)
spec.plot()
```

I then find the ratios between the original spectrum amplitude and the new amplitude, and plot that below.

```python
amps = spec.amps
amps2 = spec2.amps
ratio = amps2 / amps    
ratio[amps<560] = 0
thinkplot.plot(ratio)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim = [0,1.5], 
                     legend=False)
```

The result looks like a lot of 0's followed by a lot of 1's. I can now create an array that looks like this, and that will be my window. The results are as follows.

```python
window = numpy.ones(len(spec))
for i in range(5000):
    window[i] = 0

thinkplot.plot(window)
thinkplot.plot(ratio)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim = [0,1.5], 
                     legend=False)
```

Now I will do the same thing for a low pass filter. The only difference is that I will apply a low pass instead of a high pass filter to the spectrum of the Square wave.

```python
sig = thinkdsp.SquareSignal()
wave = sig.make_wave(duration=1, framerate=44100)
spec = wave.make_spectrum()
spec2 = wave.make_spectrum()
spec2.low_pass(5000)

amps = spec.amps
amps2 = spec2.amps
ratio = amps2 / amps    
ratio[amps<560] = 0
thinkplot.plot(ratio)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim = [0,1.5], 
                     legend=False)
```

Now our array is a lot of 1's followed by a lot of 0's. Our window is thus the following.

```python
window = numpy.zeros(len(spec))
for i in range(5000):
    window[i] = 1

thinkplot.plot(window)
thinkplot.plot(ratio)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 25000], ylim = [0,1.5], 
                     legend=False)
```

## Question 3

So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

To do this problem, I first created a square wave, and applied the filters to its spectrum. I then plotted the angles of the spectrum vs. the frequency.

```python
boxcar = numpy.ones(11)
boxcar /= sum(boxcar)
```

```python
sig = thinkdsp.SquareSignal()
wave = sig.make_wave(duration = 1, framerate = 44100)
spec = wave.make_spectrum()
spec.plot()

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude',
                     xlim=[0, 25000],  
                     legend=False)
```

```python
ys = numpy.convolve(wave.ys, boxcar, mode='same') #This applies the boxcar filter to the square wave.
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()


thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude',
                     xlim=[0, 25000],  
                     legend=False)
```

```python
angles = spectrum2.angles()
thinkplot.plot(spectrum2.fs,angles)


thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='angle (rad)',
                     xlim=[0, 25000],  
                     legend=False)
```

With the gaussian filter, we expect more phase shifts because the attenuation is not equal across all frequencies. The results below support this expectation.

```python
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
```

```python
sig = thinkdsp.SquareSignal()
wave = sig.make_wave(duration = 1, framerate = 44100)
spec = wave.make_spectrum()

ys = numpy.convolve(wave.ys, gaussian, mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum3 = smooth.make_spectrum()

angles = spectrum3.angles()
thinkplot.plot(spectrum3.fs,angles)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='angle (rad)',
                     xlim=[0, 25000],  
                     legend=False)
```

## Question 4

Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

We begin with the first window [1, -1]. I apply this window to the spectrum of a square signal to get the following.

```python
sig = thinkdsp.SquareSignal()
wave = sig.make_wave()
spectrum1 = wave.make_spectrum()
ys = numpy.convolve(wave.ys, [-1,1], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum1.plot()

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude',
                     xlim=[0, 8000],  
                     legend=False)
```

```python
spectrum2.plot()
thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude',
                     xlim=[0, 8000],  
                     legend=False)
```

This implementation of the first derivative gives us a linear function, though there is some curve. Next, I will use the window [-1/2, 2, -3/2] to get a more accurate first derivative.

```python
ys = numpy.convolve(wave.ys, [-0.5,2,-1.5], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

Thirdly, I will use the window [1,-2,1] to find the second derivative.

```python
ys = numpy.convolve(wave.ys, [1,-2,1], mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```
