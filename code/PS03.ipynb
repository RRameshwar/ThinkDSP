## Question 5

One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:

    
    Break a long signal into segments.
    Compute the DCT of each segment.
    Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
    To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.

Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
def dct_iv(ys):
    N = len(ys)
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    args = numpy.outer(ts, freqs)
    M = numpy.cos(PI2 * args)
    amps = numpy.dot(M, ys) / 2
    return amps


def DCT(wave, seg_start, num):
    total = 0
    seg = wave.segment(seg_start, wave.duration/20)
    dct = seg.make_spectrum()
    for a in range (len(dct)):
        h = dct.hs[a]
        if abs(h) < num:
            dct.hs[a] = 0               
    return dct
```

```python
def cut_amps(wave):
    total_wav = wave.segment(0,0.001)
    for i in range(20):
        seg_start = i*(wave.duration/20)
        d = DCT(wave, seg_start, 2000000)
        wav = d.make_wave()
        total_wav = total_wav.__or__(wav)
    return total_wav
```

```python
wav = thinkdsp.read_wave("100475__iluppai__saxophone-weep.wav")
wav.framerate = 2*wav.framerate
wav = wav.segment(0,5)

wav.make_spectrum().plot()
wav.make_audio()
```

```python
total_wav = cut_amps(wav)
total_wav.make_spectrum().plot()
total_wav.make_audio()
```

Within this segment, I filtered out amplitudes less than 2000000, which makes the saxophone's timbre change completely. Almost any cutoffs over this value didn't change anything, while any cutoffs less than this value cut off everything.
