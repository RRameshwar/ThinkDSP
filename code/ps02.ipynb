## Question 1

Run and listen to the examples in chap03.ipynb.  In the leakage example, try replacing the Hamming window with one of the other windows provided by NumPy, and see what effect they have on leakage.

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%precision 3
%matplotlib inline
```

We begin with creating a signal with a frequency of 440 Hz. We make two waves, one we can see clearly and one we can hear clearly. Both of these waves have a duration that's 0.5 off from a multiple of the signal's period. Therefore, we can observe leakage.

```python
signal = thinkdsp.SinSignal(freq=440)
duration_vis = signal.period * 10.5 #This is so we can see the wave.
duration_aud = signal.period * 500.5 #This is so we can hear the difference once we apply the windows.
wave_vis = signal.make_wave(duration_vis)
wave_aud = signal.make_wave(duration_aud)

wave_vis.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)

wave_aud.make_audio()
```

We can also see the leakage in the spectrum of this wave.

```python
spectrum = wave_vis.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

In order to fix the leakage, the example uses a hamming window. Instead, I write a function that applies a hanning window to the spectrum. We can see the new spectrum and then listen to the windowed wave.

```python
def hanning(wave):
    wave.ys *= numpy.hanning(len(wave.ys))
    return wave
```

```python
wave_vis = signal.make_wave(duration_vis)
wave_aud = signal.make_wave(duration_aud)

wave_vis = hanning(wave_vis)
wave_aud = hanning(wave_aud) #I apply the hanning window to both waves so we can hear and see the difference.

spectrum = wave_vis.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)

wave_aud.make_audio()
```

The hanning wave does something similar to the hamming wave with more aggressive dampening of the frequencies at the edge. In the spectrum, we can see that the amplitude is much decreased from the original, and the spiking in amplitude at the 440Hz mark is more drastic. This contrasts the gentle slope in the original spectrum.

## Question 2

Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

```python
class SawtoothChirp(thinkdsp.Chirp):
    def evaluate(self, ts):
        freq = numpy.linspace(self.start, self.end, len(ts)-1)
        dts = numpy.diff(ts)
        dps = 2*numpy.pi * freq * dts #This allows us to work with a changing set of frequencies, like in a chirp.
        
        cycles = numpy.cumsum(dps)/(2*numpy.pi)
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp) #This works with the frequencies in the same way to create a sawtooth shape.
        
        return ys

sc = SawtoothChirp(start = 440, end = 880)

wav_aud = sc.make_wave(duration = 1, framerate = 48000)
wav_vis = sc.make_wave(duration = 0.003, framerate = 48000)
wav_vis.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
wav_aud.make_audio()
```

Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
spec = wav_vis.make_spectrum()
spec.plot()
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
spec2 = wav_aud.make_spectrogram(seg_length = 1024)
spec2.plot()
thinkplot.config(xlabel='time (s)', ylabel='frequency (Hz)', ylim = [0, 10000], legend=False)
```

In both the spectrum and the spectrogram, we can observe the overtones that exist above the fundamental frequency due to aliasing. When we hear the chirp, the aliasing causes a sharp unpleasant sound in the very high frequency range.

## Question 3

In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

As the length of the trombone increases, the pitch decreases linearly. In other words, if we halve the length, we halve the pitch. If we quarter the length, the pitch is also quartered. As the pitch decreases linearly, the frequency decreases exponentially because pitch and frequency have an exponential relationship. We therefore need an ExpoChirp to immitate a glissando.

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.



```python
def glissando(low, high):
    up = thinkdsp.ExpoChirp(start = low, end = high)
    upwav = up.make_wave(duration = 1, framerate= 48000)
    down = thinkdsp.ExpoChirp(start = high, end = low)
    downwav = down.make_wave(duration = 1, framerate = 48000)
    wav = upwav.__or__(downwav)
    return wav

wav = glissando(262, 349)
wav.make_audio()    
```

The above function creates an exponential chirp going from 262Hz to 349Hz, and another that goes from 349Hz to 262Hz. Then the function concatenates the two waves and returns the glissando wave.

## Question 6

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

We begin by defining a Bartlett function that breaks the wave "wave" into segments of length M. , We then take each segment's hs and square their magnitude, then divide by M. Finally we add these magnitudes and get the average. The result is an array of complex numbers. We then plot the amplitudes of these numbers, to get a frequency vs. amplitude plot.

```python
def Bartlett(wave, M):
    total = 0
    dur = int(wave.duration)
    for i in range(0, dur):
        seg = wave.segment(i*M, M)
        mini_spec = ((seg.make_spectrum().hs)**2)/M
        total += mini_spec
    
    array = total/(dur/M)
    amp = []
    for i in array:
        amp.append(abs(i))
    
    thinkplot.plot(amp)
    thinkplot.config(title = 'Rain Spectrum Bartlett', xlabel='frequency (Hz)', ylabel='amplitude (s)', legend=False)
```

```python
Bartlett(seg, 0.1) #Runs the Bartlett method on "seg", with segments of length 0.1s
```

This plot shows us a clearer picture of frequency vs. power in the rain segment. We can see that the loudest frequencies are between 0Hz and 500 Hz, and then between 1700Hz and 2200 Hz. The overtones shown in the general spectrum above don't cut out the frequencies that we don't really hear.
